```python
import json
import re
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
list1=[]
with open('/home/midhun/Documents/python_workshop/Project/data.json') as f:
    for item in f:
        info=json.loads(item)
        list1.append(info['data']['Author'])
```

```python
list2=[]

match=re.compile(r"(\w)+(\s)?(\w)+")
for item in list1:
    result=match.search(item)
    print(result.group())
    list2.append(result.group())
```

```python
print(list2)
```

```python
from collections import Counter
```

```python
x=Counter(list2)
```

```python
x
```

```python
df=pd.DataFrame.from_dict(x,orient='index').reset_index()
```

```python
df
```

```python
large10=df.nlargest(10,0)
```

```python
large10
```

```python
plt.figure(figsize=(12,6))
x=range(10)
plt.bar(x,large10[0])
plt.xticks(x,large10['index'])

plt.xlabel('Name of people')
plt.ylabel('Number of commits')
plt.title('Top commits')

plt.show()
```
