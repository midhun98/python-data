```python
square=[i**2 for i in range(10)]
print(square)
```

```python
square=[i**2 for i in range(30)if i%3==0]
square
```

```python
square={i:i**2 for i in range(20)if i%2==0}
square
```

```python
square_rev={square[key]:key for key in square}
square_rev
```

```python
wordclean=[words.strip().lower()for words in wordlist]
```

```python
wordclean[:10]
```

```python
wordsunique=list(set(wordclean))
wordsunique
```

```python
wordsunique.sort()
```

```python
def signature(word):
    return ''.join(sorted(word))
```

```python
signature("midh un")
```

```python
def anagrams(mywords):
    return[word for word in wordclean if signature(word)==signature(mywords)]
```

```python
anagrams('cat')
```
