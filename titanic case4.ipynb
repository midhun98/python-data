```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig=plt.figure(figsize=(16,9))
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/vchrombie/da-python/master/day-4/train.csv")
```

```python
df.head()
```

```python
df.Survived.value_counts(normalize=True).plot(kind = "bar" , alpha = 0.8,color=['red','black'])
```

```python
plt.scatter(df.Survived , df.Age , alpha = 0.1,color='red')
plt.title("Age Vs Survived")
plt.show()

```

```python
df.Pclass.value_counts(normalize=True).plot(kind = "bar" , alpha = 0.6,color=['red','black','yellow'])
plt.show()
```

```python
for x in [1,2,3]:
        df.Age[df.Pclass == x].plot(kind = "kde")
plt.legend(("1st","2nd","3rd"))
plt.show()
```

```python
df.Survived[df.Sex == "male"].value_counts(normalize = True).plot(kind = "bar",color=['red','green'])
```

```python
Female_colour = "#fff000"
df.Sex[df.Survived == 1].value_counts(normalize = True).plot(kind = "bar", color = [Female_colour , 'b'])
```

```python
for x in [1,2,3]:
    df.Survived[df.Pclass == x].plot(kind = "kde")
plt.legend(("1","2","3"))
```

```python
df.Survived[(df.Sex == "male") & (df.Pclass == 1)].value_counts(normalize = True).plot(kind = "bar")
plt.title("Rich Men Survival Rate")
```

```python
df.Survived[(df.Sex == "male") & (df.Pclass == 3)].value_counts(normalize = True).plot(kind = "bar")
plt.title("Poor Men Survival Rate")
```

```python
df["hypo"] = 0
df.loc[df.Sex == "female" ,"hypo"] = 1
df["result"] = 0
df.loc[df.Survived == df.hypo, "result"] = 1
df["result"].value_counts(normalize = True)
```
